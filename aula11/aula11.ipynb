```python
%%writefile stocks-google.cu
#include <thrust/device_vector.h>
#include <thrust/host_vector.h>
#include <thrust/transform.h>
#include <thrust/reduce.h>
#include <thrust/iterator/constant_iterator.h>
#include <thrust/count.h>
#include <thrust/replace.h>
#include <iostream>
#include <fstream>
#include <chrono>  // Para medir tempos de execução
#include <cmath>   // Para funções matemáticas

// Functor para calcular (x - mean)² para o cálculo da variância
struct square_diff {
    __host__ __device__ double operator()(const double& x, const double& mean) const {
        return (x - mean) * (x - mean);  // Calcula (x - mean)²
    }
};

// Functor para calcular a diferença entre os preços de dois dias consecutivos
struct diff {
    __host__ __device__ double operator()(const double& x1, const double& x2) const {
        return x2 - x1;  // Calcula a diferença entre x2 (dia seguinte) e x1 (dia atual)
    }
};

// Functor para identificar valores positivos (dias com aumento no preço)
struct is_positive {
    __host__ __device__ bool operator()(const double& x) const {
        return x > 0;  // Retorna verdadeiro para valores positivos
    }
};

// Functor para identificar valores negativos
struct is_negative {
    __host__ __device__ bool operator()(const double& x) const {
        return x < 0;  // Retorna verdadeiro para valores negativos
    }
};

int main() {
    // Lê o arquivo stocks-google.txt
    std::ifstream file("stocks-google.txt");
    if (!file) {
        std::cerr << "Erro ao abrir o arquivo!" << std::endl;
        return 1;
    }

    // Contagem do número de linhas no arquivo
    int n = 0;
    std::string line;
    while (std::getline(file, line)) {
        n++;
    }

    file.clear();
    file.seekg(0, std::ios::beg);

    // Criação de host_vector para armazenar os preços
    thrust::host_vector<double> stocks_host(n);
    for (int i = 0; i < n; i++) {
        file >> stocks_host[i];
    }
    file.close();

    // Transferir os dados para a GPU (device_vector)
    thrust::device_vector<double> stocks_device = stocks_host;

    // -------------------- Cálculo da Variância --------------------

    auto start_variancia = std::chrono::high_resolution_clock::now();

    // Cálculo da média dos preços
    double soma = thrust::reduce(stocks_device.begin(), stocks_device.end(), 0.0, thrust::plus<double>());
    double media = soma / n;

    // Cálculo da variância usando transform com iterador constante
    thrust::device_vector<double> diff_square(n);
    thrust::transform(stocks_device.begin(), stocks_device.end(), thrust::constant_iterator<double>(media), diff_square.begin(), square_diff());

    double variancia = thrust::reduce(diff_square.begin(), diff_square.end(), 0.0, thrust::plus<double>()) / n;

    auto end_variancia = std::chrono::high_resolution_clock::now();
    std::chrono::duration<double> tempo_variancia = end_variancia - start_variancia;
    std::cout << "Variância: " << variancia << std::endl;
    std::cout << "Tempo para cálculo da variância: " << tempo_variancia.count() << " segundos" << std::endl;

    // -------------------- Análise de Variação Diária --------------------

    auto start_variacao_diaria = std::chrono::high_resolution_clock::now();

    // Vetor para armazenar as variações diárias
    thrust::device_vector<double> ganho_diario(n - 1);

    // Calcula as diferenças entre os preços consecutivos
    thrust::transform(stocks_device.begin(), stocks_device.end() - 1, stocks_device.begin() + 1, ganho_diario.begin(), diff());

    auto end_variacao_diaria = std::chrono::high_resolution_clock::now();
    std::chrono::duration<double> tempo_variacao_diaria = end_variacao_diaria - start_variacao_diaria;
    std::cout << "Tempo para cálculo da variação diária: " << tempo_variacao_diaria.count() << " segundos" << std::endl;

    // -------------------- Contagem de Dias com Aumento no Preço --------------------

    auto start_contagem_aumento = std::chrono::high_resolution_clock::now();

    // Contagem de dias com aumento no preço das ações
    int dias_com_aumento = thrust::count_if(ganho_diario.begin(), ganho_diario.end(), is_positive());

    auto end_contagem_aumento = std::chrono::high_resolution_clock::now();
    std::chrono::duration<double> tempo_contagem_aumento = end_contagem_aumento - start_contagem_aumento;
    std::cout << "Número de dias com aumento no preço: " << dias_com_aumento << std::endl;
    std::cout << "Tempo para contagem dos dias com aumento: " << tempo_contagem_aumento.count() << " segundos" << std::endl;

    // -------------------- Cálculo do Aumento Médio nos Dias de Aumento --------------------

    auto start_aumento_medio = std::chrono::high_resolution_clock::now();

    // Substituir valores negativos por zero
    thrust::replace_if(ganho_diario.begin(), ganho_diario.end(), is_negative(), 0.0);

    // Soma dos aumentos nos dias positivos
    double soma_aumentos = thrust::reduce(ganho_diario.begin(), ganho_diario.end(), 0.0, thrust::plus<double>());

    // Cálculo do aumento médio
    double aumento_medio = dias_com_aumento > 0 ? soma_aumentos / dias_com_aumento : 0.0;

    auto end_aumento_medio = std::chrono::high_resolution_clock::now();
    std::chrono::duration<double> tempo_aumento_medio = end_aumento_medio - start_aumento_medio;
    std::cout << "Aumento médio nos dias em que o preço subiu: " << aumento_medio << std::endl;
    std::cout << "Tempo para cálculo do aumento médio: " << tempo_aumento_medio.count() << " segundos" << std::endl;

    return 0;
}
```

Esses resultados destacam a vantagem de usar a GPU para tarefas que envolvem grandes volumes de dados e cálculos repetitivos, como transformações ponto a ponto e operações de redução, especialmente no contexto de análise de séries temporais, como as de preços de ações.
